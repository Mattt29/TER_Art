```python
from bokeh.io import show
from bokeh.models import Panel, Tabs, ColumDataSource
from bokeh.plotting import figure

plot_figure1 = figure(
        title='UMAP projection of the dataset',
        plot_width=900,
        plot_height=600,
        tools=('pan, wheel_zoom, reset')
    )
plot_figure1.circle(
        'x',
        'y',
        source=datasource1)

tab1 = Panel(child=plot_figure1, title="UMAP")

plot_figure2 = figure(
    title='PCA projection of the dataset',
    width=900,
    height=600,
    tools=('pan, wheel_zoom, reset'))
plot_figure2.circle(
    'x',
    'y',
    source=datasource2)

tab2 = Panel(child=plot_figure2, title="PCA")

show(Tabs(tabs=[tab1, tab2]))
```
